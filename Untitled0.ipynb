```python
from covid19dh import covid19
from datetime import datetime
import matplotlib.pyplot as plt
```

```python
oct21USA, srcUSA = covid19("USA", start = datetime(2021,10,1), end = "2021-10-31", raw='false' )
```

```python
oct21CAN, srcCAN = covid19("CAN", start = datetime(2021,10,1), end = "2021-10-31", raw='False')
```

```python
fig, axs= plt.subplots(4,2,figsize=(10,10))
x= oct21USA['date']
yValues=['tests','hosp','icu','deaths']

for row in range(0,4):
  for col in range(0,2):
    y=oct21USA[yValues[row]] if col==0 else oct21CAN[yValues[row]]
    yTitle='USA'+' '+yValues[row] if col==0 else 'canada'+' '+yValues[row]
    axs[row,col].plot(x,y)
    axs[row,col].set_title(yTitle)
    fig.tight_layout()
```
